```python
# LT-method dtdW
import numpy as np
import matplotlib.pyplot as plt

T=1;b=2;sigma=1
mc=10000
Err_strong=[]
Err_weak=[]

Dt=[2**(-10),2**(-9),2**(-8),2**(-7)]

for k in Dt:
    N=int(T/k)
    np.random.seed(mc)
    Winc=np.sqrt(k)*np.random.randn(mc,N)
    B=np.cumsum(Winc,axis=1)
    X_true=np.exp((b-0.5*sigma**2)*T*np.ones(mc)+ sigma*B[:,-1])
    X_LT=np.ones(mc)
    for j in range(N):
        X_LT+= b*X_LT*k
        X_LT+= sigma*X_LT*Winc[:,j]
    X= abs(X_true- X_LT)
    Y= abs(np.mean(X_LT,0)-np.mean(X_true,0))
    Err_strong.append(np.mean(X,0))
    Err_weak.append(Y)
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.linear_model import LinearRegression
```

```python
h = np.array(Dt).reshape(-1,1)
reg1 = LinearRegression().fit(np.log(h), np.log(Err_strong))
reg2 = LinearRegression().fit(np.log(h), np.log(Err_weak))
```

```python
pal = ["#FBB4AE","#B3CDE3", "#CCEBC5","#CFCCC4"]
plt.loglog(Dt , Err_strong , 'P')
plt.loglog(Dt, Err_weak, 'D')

x=np.linspace(0.0001,0.08)
#y=(reg1.coef_[0])*x + reg1.intercept_
y = np.exp(reg1.intercept_) * x**reg1.coef_
z = np.exp(reg2.intercept_) * x**reg2.coef_
s = np.exp(1) * x**(0.5)
r = np.exp(0.5) * x**(1)
plt.loglog(x, y, color=pal[2], label='LT strong')
plt.loglog(x, z, color=pal[3], label='LT weak')
plt.loglog(x, s, 'g-.' , color='brown', label='Order 0.5')
plt.loglog(x, r, ':' , color='red', label= 'Order 1')


plt.title('LT Method')
plt.xlabel('$\Delta t$'); plt.ylabel('Error'); plt.legend(loc=4);

plt.show()      
```
